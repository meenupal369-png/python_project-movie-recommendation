```python
import ast
def fetch_director(obj):
    L=[]
    try:
        for i in ast.literal_eval(obj):
            if i['job']=='Director':
                L.append(i['name'])
                break
    except (SyntaxError, ValueError):
        pass  
    return L

```

```python
movies['crew'].apply(fetch_director)
```

```python
import ast
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
           L.append(i['name'])
           counter=counter+1
        else:
            break
    return L
```
